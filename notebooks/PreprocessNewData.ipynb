```python
out_dir = "/home/yyelisieiev/luftr_data/add_rivets"
os.mkdir(out_dir)
```

```python
for index in range(len(df)):
    row = df.iloc[index]
    image_path = os.path.join(data_path, row[5])
    x, y, w, h = row[1:5]
    img = np.array(Image.open(image_path))
    rivet = Image.fromarray(img[y: y + h, x: x + w])
    file_name = f"{index}_rivet.png"
    rivet.save(os.path.join(out_dir, file_name))    
```
